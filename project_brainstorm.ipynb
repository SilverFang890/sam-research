```python
bitcoin = Table().read_table('./archive/Bitcoin.csv')
```

```python
btc = bitcoin.with_column('number', np.arange(bitcoin.num_rows, 0, -1))
btc = btc.move_column('number', 0)
```

```python
btc.scatter(0, 2)
plots.show()
```

```python
def visualize(coin):
    vizcoin = coin.with_column('number', np.arange(coin.num_rows, 0, -1))
    vizcoin = vizcoin.move_column('number', 0)
    vizcoin.scatter(0, 2)
    plots.show()
```

```python
dogecoin = Table().read_table('./archive/Meme Coin/Dogecoin.csv')
visualize(dogecoin)
```

```python
ethereum = Table().read_table('./archive/Ethereum.csv')
visualize(ethereum)
```
